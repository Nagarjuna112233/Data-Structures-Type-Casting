# Discretization 
**apply discretization using intervals**

```python
df['age_category']= pd.cut(df['age'],
                           bins = [0,12,19,30,60,90],
                           labels=['child','teenager','young_adult',
                                   'middle_aged','senior_citizen'])                
```

```python
df[['age','age_category']]
```

```python
pd.crosstab(df['age_category'],df['stroke'],margins= True)
```

```python
df.groupby("stroke")['age'].describe().transpose()
```
