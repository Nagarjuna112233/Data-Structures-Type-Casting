# Filtering

- Selecting/extracting the data based on condition(s)


**using 1 condition**

```python
df[df["gender"]=="Male"]
```

```python
df[(df['tip']>=6)]
```

```python
df[(df['no_of_members']>=6)]
```

```python
df[(df['time']>='Lunch')]
```

```python
df[(df['total_bill']>=40)]
```

```python
df[(df['tip']<6)]
```

```python
df[(df['smoker']>='Yes')]
```

**using multiple conditions**
- Syntax : df[(condition1) or/and (condtion2)]

```python
df[(df['tip']>=7 & (df["gender"]=="Male"))]
```

```python
df.info()
```

```python
df[(df['tip']>=3) | (df['gender']=='Female')]
```

```python
df[(df['gender']=='Female') | (df['tip']>=3) & (df['total_bill']<=60)]
```

```python
df[(df['gender']=='Female')  & (df['total_bill']>=25)]
```
