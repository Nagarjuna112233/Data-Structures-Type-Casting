# Joint Plot

A join plot allows to study the relationship between 2 numeric varaiables. The central chart display their correlation.it is usually a scatterplot, a hexbin plot, a 2D histogram or a 2d density plot. 

```python
sns.jointplot(x ='total_bill',y = 'tip' ,data = df)
plt.show()
              
```

# Violon Plot

Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x ='total_bill',y = 'sex',data=df)
plt.show()
```

# Pair Plot-Multiple continous variables

A "Pairs Plot" is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value.like this: pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df,vars=continous)
plt.show()
```
